```python
# Importing libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt, floor
import seaborn as sn

points = np.array([[10, 11, 11],
                   [12, 9, 9],
                   [8, 7, 8],
                   [0, 0, 0],
                   [-2, 4, 4],
                   [2, -4, 3],
                   [3, 2, -1],
                   [7, 3, 9],
                   [8, 4, 5],
                   [3, 5, 6]])
                  
n, _ = points.shape
clusters = [[points[x,:]] for x in range(n)]
```

```python
def dis(x, y):
    return np.sqrt(np.sum(np.power(np.subtract(x, y), 2)))

def centroid(clus):
    n = len(clus)
    if n == 1:
        return clus[0]
    else:
        g = []
        for i in range(n):
            g.append(centroid(clus[i]))
        G = np.zeros(3)
        for i in range(n):
            G += g[i]
        return G/n
        
'''
clus_dis(clus1, clus2, mode)
mode:
single-link-method : 1
𝐶𝑜𝑚𝑝𝑙𝑒𝑡𝑒−𝑙𝑖𝑛𝑘 𝑚𝑒𝑡ℎ𝑜𝑑 : 2
𝐴𝑣𝑒𝑟𝑎𝑔𝑒−𝑙𝑖𝑛𝑘 𝑚𝑒𝑡ℎ𝑜𝑑 : 3
𝐶𝑒𝑛𝑡𝑟𝑜𝑖𝑑 𝑚𝑒𝑡ℎ𝑜𝑑 : 4
'''  
def clus_dis(clus1, clus2, mode):
    if isinstance(clus1[0], np.ndarray) and isinstance(clus2[0], np.ndarray):
        if mode < 5:
            return dis(clus1[0], clus2[0])
        else:
            print('Invalid mode!')
            return
    else:
        if isinstance(clus1[0], np.ndarray):
            n1 = 1
            n2 = len(clus2)
        elif isinstance(clus2[0], np.ndarray):
            n2 = 1
            n1 = len(clus1)
        else:
            n1 = len(clus1)
            n2 = len(clus2)
        D = np.zeros(n1*n2)
        r = 0
        for i in range(n1):
            for j in range(n2):
                if isinstance(clus1[0], np.ndarray):
                    D[r] = clus_dis(clus1, clus2[j], mode)
                elif isinstance(clus2[0], np.ndarray):
                    D[r] = clus_dis(clus1[i], clus2, mode)
                else:
                    D[r] = clus_dis(clus1[i], clus2[j], mode)
                r += 1
        if mode == 1:
            return np.min(D)
        elif mode == 2:
            return np.max(D)
        elif mode == 3:
            return np.sum(D)/(n1*n2)
        elif mode == 4:
            G1 = centroid(clus1)
            G2 = centroid(clus2)
            return dis(G1, G2)
    
    
def Agnes(clusters, mode):
    num = len(clusters)
    if num == 1:
        print("Error! There is only 1 cluster!")
        return
    else:
        tup = int(num*(num-1)/2)
        distants = np.zeros(tup)
        i1 = np.zeros(tup)
        i2 = np.zeros(tup)
        r = 0
        for i in range(num-1):
            for j in range(i+1,num):
                i1[r] = i
                i2[r] = j
                distants[r] = clus_dis(clusters[i], clusters[j], mode)
                r += 1
        
        m = np.argmin(distants)
        index1 = int(i1[m])
        index2 = int(i2[m])
        print(f"{index1} and {index2}")
        n1 = len(clusters[index1])
        n2 = len(clusters[index2])
        new = [clusters[index1], clusters[index2]]
        clusters[index1] = new
        clusters.pop(index2)
        return clusters, distants[m]
            
    
```

```python
mode = 1
clusters = [[points[x,:]] for x in range(n)]
for i in range(n-1):
    clusters, diss = Agnes(clusters, mode)
    print(f"step {i+1}, length={len(clusters)}, distance = {diss}\n")
    for clus in clusters:
        print(clus)
    print('\n')
```
