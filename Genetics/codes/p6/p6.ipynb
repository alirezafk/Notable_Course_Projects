```python
# Importing libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
import random
```

```python
def random_str(n):
    chars = string.ascii_letters + ' '
    return random.sample(chars, n)
```

```python
class Member:
    def __init__(self, mem_size = None, guess = None):
        self.mem_size = mem_size
        self.score = 0.01
        if guess:
            self.guess = guess
            self.mem_size = len(guess)
        else:
            self.guess = random_str(mem_size)
    
    def  __str__(self):
        return "".join(self.guess)
    
    def calc_score(self, password):
        self.score = max(self.comparison(password), 0.01)
        
    def comparison(self, different_guess):
        a = 0
        for x,y in zip(self.guess, different_guess):
            if x == y:
                a += 1
        return a/self.mem_size
```

```python
class pop:
    def __init__(self, pop_size, mem_size, password = "you can crack it", mut_prob = 0.01):
        self.mut_prob = mut_prob
        self.pop_size = pop_size
        self.mem_size = mem_size
        self.members = [Member(mem_size) for x in range(pop_size)]
        self.password = password
    
    # A method to choose parents according to their probabilities
    def choose_parents(self):
        T = 0
        for x in self.members:
            x.calc_score(self.password)
            T += x.score
        probabilities = []
        members = []
        for mem in self.members:
            mem.score = mem.score/T
            probabilities.append(mem.score)
            members.append(mem)
        for i in range(self.pop_size):
            yield np.random.choice(members, 2, p = probabilities)
    
    # A method to create new member from two parents
    def create_new_member(self, parent1, parent2):
        # choosing a random index to cut parent's guesses
        cut_index = np.random.choice(len(parent1.guess), 1)[0]
        # crossover
        part1 = parent1.guess[:cut_index]
        part2 = parent2.guess[cut_index:]
        part1.extend(part2)
        child = part1
        return child
    
    def new_generation(self):
        generation = []
        for parent1, parent2 in self.choose_parents():
            # crossovering
            child = self.create_new_member(parent1, parent2)
            # mutation
            for i in range(len(child)):
                if np.random.random() <= self.mut_prob:
                    child[i] = random_str(1)[0]
            generation.append(Member(guess = child))
        self.members = list(generation)
    
    def stop(self):
        for i, x in enumerate(self.members):
            if str(x) == self.password:
                return i
        return False
    
```

```python
password = "My name is Alireza"
population = pop(300, len(password), password)
n = 0
while(population.stop() == False):
    population.new_generation()
    n += 1
index = population.stop()
print(f"The number of generations = {n}")
print(f"First index of correct guess = {index}")
print(f"The {index}'th member of population is = {population.members[index]}")
```
